```python
data = pd.read_csv("netflix1.csv")
```

```python
# Deleting redundant columns
data.head(1)
```

```python
data.columns
```

```python
data.drop(columns= "rating", inplace = True)
```

```python
data.columns
```

```python
# Renaming the columns, capitalize
new_column_names = []

for i in data.columns:
    new_column_names.append(i.capitalize())
```

```python
new_column_names
```

```python
data.columns = new_column_names
```

```python
data.columns
```

```python
# Dropping duplicates: data.drop_duplicates(inplace = True)
data.duplicated().sum()
```

```python
# Remove Nan values from dataset: data.dropna(inplace = True)
data.isna().sum()
```

```python
# Check for more transformations
data.head()
```

```python
data["Show_id"].unique()
```

```python
data["Show_id"] = data["Show_id"].apply(lambda x: x.split("s")[1])
```

```python
data
```

```python
type(data["Show_id"][0])
```

```python
data["Show_Id"] = data["Show_id"].astype(int)
```

```python
data
```

```python
# change date format from / to -
data["Date_added"] = data["Date_added"].apply(lambda x: x.replace("/","-"))
```

```python
data.to_csv("Cleaned_Data_Netflix.csv", index = False)
```
